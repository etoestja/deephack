```python
import gym
env = gym.make('MsPacman-v0')
observation = env.reset()
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
obs = []
for i in range(100):
    action = env.action_space.sample()
    observation, reward, done, info = env.step(action)
    obs.append(observation)
```

```python
obs = np.array(obs)
```

```python
#frame = _process_frame42(observation)
#frame = cv2.resize(frame, (105, 80))
frame = observation - obs[50]
plt.imshow(frame)
```

```python
frame = frame.mean(2)
fft = np.fft.fft2(frame, norm='ortho')
```
